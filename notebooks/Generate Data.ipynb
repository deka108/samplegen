## Random

```python
def random_dist(data, N):
    return np.random.choice(data, size=N)
```

```python
fig = plt.figure(figsize=(15, 5))
sns.countplot(random_dist([0,1], 20))
```

```python
fig = plt.figure(figsize=(15, 5))
sns.countplot(random_dist(np.arange(0, N), 20))
```

```python
fig = plt.figure(figsize=(15, 5))
sns.countplot(random_dist(np.arange(0, N), 100))
```

```python
sns.distplot(random_dist(np.arange(N), 1000), hist=True)
```

## Random Uniform

```python
def random_uniform_dist(data, N):
    data_size = len(data)
    p = 1/data_size
    probs = [p for _ in range(data_size)]
    
    return np.random.choice(data, size=N, p=probs)
```

```python
fig = plt.figure(figsize=(15, 5))
sns.countplot(random_uniform_dist([0,1], 50))
```

```python
fig = plt.figure(figsize=(15, 5))
sns.countplot(random_uniform_dist(np.arange(0, N), 20))
plt.savefig("../docs/random_uniform.png")
```

```python
plt.figure(figsize=(15, 5))
sns.countplot(random_uniform_dist(np.arange(0, N), 100))
```

```python
sns.distplot(random_uniform_dist(np.arange(N), 1000), hist=True)
plt.savefig("../docs/random_uniform_hist.png")
```

## Random Normal

```python
def random_normal_dist(data, N, left=0.01, right=0.99):
    data_size = len(data)
    x = np.linspace(ss.norm.ppf(left), ss.norm.ppf(right), data_size)
    probs = (ss.norm.pdf(x)/ss.norm.pdf(x).sum())
    
    return np.random.choice(data, size=N, p=probs)
```

```python
# uniformly distributed data
fig = plt.figure(figsize=(15, 5))
sns.countplot(random_normal_dist(np.arange(N), 20))
```

```python
# uniformly distributed data
fig = plt.figure(figsize=(15, 5))
sns.countplot(random_normal_dist(np.arange(N), 100))
```

```python
sns.distplot(random_normal_dist(np.arange(N), 1000), hist=True)
plt.savefig("../docs/random_normal_hist.png")
```

## Skewed Left

```python
def left_skew_dist(data, N, a=2, loc=2, left=0.01, right=0.99):
    data_size = len(data)
    x = np.linspace(ss.gamma.ppf(left, a, loc=loc), ss.gamma.ppf(right, a, loc=loc), data_size)
    
    probs = (ss.gamma.pdf(x, a)/ss.gamma.pdf(x, a).sum())
    
    return np.random.choice(data, size=N, p=probs)
```

```python
fig = plt.figure(figsize=(15, 5))
sns.countplot(left_skew_dist(np.arange(N), 20))
```

```python
fig = plt.figure(figsize=(15, 5))
sns.countplot(left_skew_dist(np.arange(N), 100))
```

```python
sns.distplot(left_skew_dist(np.arange(N), 1000), hist=True)
plt.savefig("../docs/left_skew_dist.png")
```

## Skewed Right

```python
def right_skew_dist(data, N, a=2, loc=2, left=0.01, right=0.99):
    data_size = len(data)
    x = np.linspace(ss.gamma.ppf(left, a, loc=loc), ss.gamma.ppf(right, a, loc=loc), data_size)
    
    probs = (ss.gamma.pdf(x, a)/ss.gamma.pdf(x, a).sum())[::-1]
    
    return np.random.choice(data, size=N, p=probs)
```

```python
fig = plt.figure(figsize=(15, 5))
sns.countplot(right_skew_dist(np.arange(N), 20))
```

```python
fig = plt.figure(figsize=(15, 5))
sns.countplot(right_skew_dist(np.arange(N), 100))
```

```python
sns.distplot(right_skew_dist(np.arange(N), 1000), hist=True)
plt.savefig("../docs/right_skew_dist.png")
```
